```python
import numpy as np
import matplotlib.pyplot as plt
import requests 
import plotly.express as px
import pandas as pd
```

```python
url = 'http://191.252.218.131:8000/index.php'
r = requests.get(url)
print("Status code:", r.text)
```

```python
data = r.text.split("<br><br>")
```

```python
data_Te1=[]
data_Te2=[]
data_Te3=[]
data_Cur=[]
index = []
for i,line in enumerate(data):
  aux=line.split(" ")
#  print(i,aux)
  if (len(aux)>=6):
    index.append(float(i))
    data_Te1.append(float(aux[3]))
    data_Te2.append(float(aux[4]))
    data_Te3.append(float(aux[5]))
    data_Cur.append(float(aux[6]))
```

```python
print(data_Te1)
```

```python
plt.plot(data_Te1)
plt.plot(data_Te2)
plt.plot(data_Te3)
plt.plot(data_Cur)
plt.show()

```

```python
data = {'x':index,'T1':data_Te1,'T2':data_Te2,'T3':data_Te3,'Cur':data_Cur}
df = pd.DataFrame(data)
```

```python
fig = px.line(data_frame=df, x="x", y=["T1","T2","Cur"], width=1000, height=800)
fig.show()
```
